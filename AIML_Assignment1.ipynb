```python
#Generate a Data set for Health Care using the following:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
people=['ram', 'hanuman','krishn','dharma','eshwar','mani','madhu','dixit','tejas']
age=[78,47,89,50,34,20,19,11,90,]
height=[160,135,165,173,168,59,105,155,158]
weight=[30,44,62,38,55,64,65,32,91,]

```

Scatter plot between height and weight

```python

plt.scatter(height,weight,plt.grid(True))
plt.title("scatter plot between heigth and weight")
plt.xlabel("height ")
plt.ylabel("weight")
plt.show()
```

**Bar Graph bettween people and weight**

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(people,weight,width=0.6)
plt.title("Bar Graph bettween people and there height")
plt.xlabel("people")
plt.xticks(fontsize=14,rotation=30)
plt.ylabel("height")
plt.yticks(fontsize=14)
plt.show()
```

Generate Histogram

```python
plt.hist(age,bins=5,edgecolor='k',color='g')
plt.title("Histogram of patient age")
plt.xlabel("age")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
